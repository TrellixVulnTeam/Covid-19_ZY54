En esta notebook, se implementa: 
- **Extract**: Extraccion de los ficheros de tipos csv, de las diferentes fuentes.
- **Transform** : Transformacion de los datos que estan en los ficheros csv, asi como incorporar nuevos datos utilies.
- **Load** : En este repositorio, no se cargara a ninguna Data lake o Warehouse, solo dejaremos los datos preparados para su analisis.

```python
#Importando las libreerias a usar
import pandas as pd
import numpy as np
import json
```

# **Extract**

```python
df_nacional = pd.read_csv('../../datasets/Casos_covid.csv')
df_municipal = pd.read_csv('../../datasets/Casos_Diarios_Municipio_.csv')
```

```python
#Datos nivel nacional
df_nacional.head(5)
```

```python
#Datos por municipio
df_municipal.head(5)
```

```python
#Coordenadas de  estados
with open('./../../datasets/cord.json', 'r', encoding='utf-8') as file:
    coordenada = file.read()

coordenada = json.loads(coordenada)
coordenada
```

# Transform

```python
""" Llamamos a la funcion para generar dataframe """
%run './../funciones/Function_where.ipynb'
```

```python
""" Cambiando de nombre una columna"""
df_nacional.rename(columns={'nombre':'Estado'}, inplace=True)
df_nacional.head(3)
```

```python
df_nacional.head(3)
```

```python
""" Pasando a mayusculas todas los estados """
df_coordenada = pd.DataFrame(coordenada)
estado = [ w.upper() for w in df_coordenada.Estado]
df_coordenada['Estado'] = estado
df_coordenada.head(4)
```

```python
""" Probando una merge con 2 tablas """
pd.concat([df_coordenada.sort_values(by='Estado'), df_nacional.sort_values(by='Estado')], axis=1)
```

```python
pd.merge(df_coordenada.sort_values(by='Estado'), df_nacional.sort_values(by='Estado'))
```

```python
""" Guardando el merge de 2 tablas"""
df_completo = pd.merge(df_coordenada.sort_values(by='Estado'), df_nacional.sort_values(by='Estado'))
df_completo.head(4)
```

```python
"""Total de datos faltantes"""
df_completo.isna().sum()
```

# LOAD

#### En la carga, solo se guardara en un fichero de tipo csv, aunque podria usarse un data warehouse. Pero para el fin del repositorio no sera tan necesario.
