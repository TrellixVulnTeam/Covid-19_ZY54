## Seleccionando Casos Psitivos

```python
df_positivo = df[df["RESULTADO"] == "Positivo"]
```

```python
df_positivo = df_positivo[["FECHA_INGRESO", "RESULTADO"]]
```

```python
df_positivo.head()
```

```python
df_positivo["FECHA_INGRESO"].value_counts()
```

```python
df_positivo.columns = ['Fecha', 'Casos']
```

```python
#Realizando la suma de los casos
casos = df_positivo["Fecha"].value_counts()
#Conviertiedo a DataFrame
df_positivos = pd.DataFrame(casos)
#Reseteando el index
df_positivos = df_positivos.reset_index()
```

```python
#Renombrando columnas
df_positivos.columns = ['Fecha', 'Casos']
```
