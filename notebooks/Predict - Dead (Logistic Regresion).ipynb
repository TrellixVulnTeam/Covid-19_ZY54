```python
df["EMBARAZO"] = np.where(df["EMBARAZO"] == 97, 0, df["EMBARAZO"])
df["EMBARAZO"] = np.where(df["EMBARAZO"] == 2, 0, df["EMBARAZO"])
df["EMBARAZO"] = np.where(df["EMBARAZO"] == 1, 1, df["EMBARAZO"])
df["EMBARAZO"] = np.where(df["EMBARAZO"] == 98, 0, df["EMBARAZO"])
```

```python
def Chage_var(df, column, var_actual, var_new):
    return np.where(df[column] == var_actual, var_new, df[column])
```

```python
df.head(3)
```

```python
df['NEUMONIA'] = Chage_var(df, 'NEUMONIA', 99, 0)
```

```python
df['NEUMONIA'] = Chage_var(df, 'NEUMONIA', 2, 0)
```
