```python
result = seasonal_decompose(df_positivos["Casos"], model = 'additive')
result.plot()
plt.show()
```

### Multiplicativo

```python
dec_multiplicative = seasonal_decompose(df_positivos.Casos , model = 'multiplicative')
dec_multiplicative.plot()
plt.show()
```

### Suavisado

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
span = 12
alpha = 2/(span+1)

df_positivos["E12"] = df_positivos["Casos"].ewm(alpha=alpha, adjust=False).mean()
df_positivos["S12"] = SimpleExpSmoothing(df_positivos["Casos"]).fit(smoothing_level=alpha, optimized=False).fittedvalues.shift(-1)
df_positivos.head()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df_positivos["DESadd12"] = ExponentialSmoothing(df_positivos["Casos"], trend='add').fit().fittedvalues.shift(-1)
df_positivos.head()
```

```python
df_positivos[["Casos", "E12", "DESadd12"]].iloc[:50].plot(figsize=(20,6)).autoscale(axis='x', tight=True);
```
